# Preprocessing Pipeline for the Noised Data

```python
import os
import cv2
from pathlib import Path
from tqdm import tqdm
```

```python
NOISED_ROOT = "../dataset/noised_test"
PROCESSED_ROOT = "../dataset/noised_test_processed"

IMG_SIZE = 224
IMG_EXTS = {'.jpg'}
```

```python
def preprocess_image(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = cv2.medianBlur(img, 3)
    return img
```

```python
for variant in os.listdir(NOISED_ROOT):
    variant_path = os.path.join(NOISED_ROOT, variant)
    if not os.path.isdir(variant_path):
        continue

    for cls in os.listdir(variant_path):
        cls_path = os.path.join(variant_path, cls)
        if not os.path.isdir(cls_path):
            continue

        out_dir  = os.path.join(PROCESSED_ROOT, variant, cls)
        os.makedirs(out_dir, exist_ok=True)

print("Folders Created Successfully\n")
```

```python
for variant in os.listdir(NOISED_ROOT):
    variant_path = os.path.join(NOISED_ROOT, variant)
    if not os.path.isdir(variant_path):
        continue

    for cls in os.listdir(variant_path):
        cls_path = os.path.join(variant_path, cls)
        if not os.path.isdir(cls_path):
            continue

        images = [f for f in os.listdir(cls_path) if os.path.splitext(f)[1].lower() in IMG_EXTS ]
        out_dir = os.path.join(PROCESSED_ROOT, variant, cls)

        for img_name in tqdm(images, desc=f"{variant}-{cls}"):
            src = os.path.join(cls_path, img_name)
            dst = os.path.join(out_dir, img_name)
            img = preprocess_image(src)
            cv2.imwrite(dst, img)

print("\nFolders Created Successfully\n")
```
