# Preprocessing Pipeline 

```python
import os
import cv2
import shutil
import random
from pathlib import Path
from tqdm import tqdm
import numpy as np 

SEED = 67
random.seed(SEED)
np.random.seed(SEED)
```

```python
RAW_PATH = "../dataset/raw/L1"
OUT_PATH = "../dataset/processed"
RAW_OUT_PATH = "../dataset/raw_split"

IMG_SIZE = 224
TRAIN_RATIO = 0.75
VAL_RATIO = 0.05
TEST_RATIO = 0.20
```

```python
splits = ["train", "val", "test"]

for s in splits:
    os.makedirs(os.path.join(OUT_PATH, s), exist_ok=True)
    os.makedirs(os.path.join(RAW_OUT_PATH, s), exist_ok=True)

print("Folders created")
```

```python
def preprocess_image(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = cv2.medianBlur(img, 3)
    return img
```

```python
classes = os.listdir(RAW_PATH)

for cls in classes:
    cls_path = os.path.join(RAW_PATH, cls)
    images = os.listdir(cls_path)
    random.shuffle(images)

    n = len(images)
    n_train = int(n * TRAIN_RATIO)
    n_val = int(n * VAL_RATIO)

    train_imgs = images[:n_train]
    val_imgs = images[n_train:n_train+n_val]
    test_imgs = images[n_train+n_val:]

    split_data = {"train": train_imgs, "val": val_imgs, "test": test_imgs}

    for split in split_data:
        proc_dir = os.path.join(OUT_PATH, split, cls)
        raw_dir  = os.path.join(RAW_OUT_PATH, split, cls)
        os.makedirs(proc_dir, exist_ok=True)
        os.makedirs(raw_dir,  exist_ok=True)

        for img_name in tqdm(split_data[split], desc=f"{cls}-{split}"):
            src = os.path.join(cls_path, img_name)
            
            dst_proc = os.path.join(proc_dir, img_name)
            img = preprocess_image(src)
            cv2.imwrite(dst_proc, img)

            dst_raw = os.path.join(raw_dir, img_name)
            shutil.copy2(src, dst_raw)

print("\nFolders Created Successfully \n")
```
